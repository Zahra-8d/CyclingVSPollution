```python
# Specify the directory path
directory_path = '/home/yuna/github/CyclingVSPollution/yuna/project_data/London_geometry/'

# Get a list of all files in the directory
files = os.listdir(directory_path)

# Filter only .shp files (excluding directories)
shp_files = [f for f in files if f.endswith('.shp') and os.path.isfile(os.path.join(directory_path, f))]
shp_files
```

```python
# Load the shapefile 
def load_shp_file(shp_file_path):
    gdf = gpd.read_file(shp_file_path)
    return gdf
```

```python
# get the name of the borough from shp
def get_borough_name(shp_file_path):
    borough_name = shp_file_path.split('/')[-1].replace('.shp',"")
    return borough_name
```

```python
# Add the extracted borough name as a new column
def add_column_borough(borough_name):
    gdf['Borough_Name'] = borough_name
    return gdf
```

```python
# Find rows containing the borough name and update the name
def add_update_borough(df):
    df.loc[df.apply(lambda row: row.astype(str).str.contains(borough_name, case=False, na=False).any(), axis=1), 'Borough_Name'] = borough_name
    return df
```
