```python
import pandas as pd
import json
```

```python
# File paths
csv_file_path = '/home/yuna/github/CyclingVSPollution/yuna/project_data/london_with_borough_name.csv'
geojson_file_path = '/home/yuna/github/CyclingVSPollution/yuna/project_data/london_boroughs.geojson'
output_csv_file_path = 'london_boroughs_with_geometry.csv'
final_path = 'london_NO2_borough.csv'
```

```python
# Load CSV data
csv_data = pd.read_csv(csv_file_path)
```

```python
# Load GeoJSON data
with open(geojson_file_path, 'r') as f:
    geojson_data = json.load(f)
```

```python
# Extract borough names and geometries from GeoJSON
geojson_boroughs = {
    feature['properties']['name']: feature['geometry']
    for feature in geojson_data['features']
}
```

```python
# Add geometry to CSV data based on matching borough names
csv_data['geometry'] = csv_data['borough_name'].map(geojson_boroughs)
```

```python
csv_final = csv_data[['borough_name', 'Average concentration roadside*', 'geometry']]
csv_final
```

```python
# Save the updated CSV file with geometry column
csv_final.to_csv(final_path, index=False)
```
